```python
# Checking the Id proof/Documents submitted value counts against target Variable .
# Checking for records only where there is payment difficulty i.e target = 1

DocumentsColumnList  = []
for i in difficulties_in_payment.columns:
    if i.startswith('ad_flag_document'):
        DocumentsColumnList.append(i)


DocumentFlag = pd.DataFrame()

for i in DocumentsColumnList:
    Scheck = difficulties_in_payment[i].value_counts(normalize=True)
    DocumentFlag.loc[:, i] = Scheck

DocumentFlag
```

```python
# Based on the above output the Flag Document columns can be dropped as
# almost all the FlagDocumentscolumns have value as 0 or 1 for nearly 99% of data. There is no clear information about the type of documents.

dropColumnList.extend(DocumentsColumnList)
```

```python
# Checking the Contact Details  value counts against target Variable .
# Checking for records only where there is payment difficulty i.e target = 1

ContactFlagColumnsList = ['ad_flag_mobil_ct',
'ad_flag_emp_phone_ct',
'ad_flag_work_phone_ct',
'ad_flag_cont_mobile_ct',
'ad_flag_phone_ct',
'ad_flag_email_ct'
]

ContactFlag = pd.DataFrame()

for i in ContactFlagColumnsList:
    Scheck = difficulties_in_payment[i].value_counts(normalize=True)
    ContactFlag.loc[:, i] = Scheck

ContactFlag
```

```python

zscorethreshold = 3
columnnames = ['ad_sk_id_prev _ct', 'ad_sk_id_curr_ct', 'ad_amt_annuity_ct',
       'ad_amt_application_ct', 'ad_amt_credit_ct', 'ad_amt_down_payment_ct',
       'ad_amt_goods_price_ct', 'ad_hour_appr_process_start_ct',
       'ad_nflag_last_appl_in_day_ct', 'ad_rate_down_payment_ct',
       'ad_days_decision_ct', 'ad_sellerplace_area_ct', 'ad_cnt_payment_ct',
       'ad_days_first_drawing_ct', 'ad_days_first_due_ct',
       'ad_days_last_due_1st_version_ct', 'ad_days_last_due_ct',
       'ad_days_termination_ct', 'ad_nflag_insured_on_approval_ct']
for columnname in columnnames:
    values = previous_data1[columnname]


    if not pd.api.types.is_numeric_dtype(values):
        continue

    zscores = np.abs((values - values.mean()) / values.std())

    aberrantz = values[zscores > zscorethreshold]
    zaberrantpercentage = len(aberrantz) / len(values) * 100

    print(f"Column: {columnname}")
    print(f"Z-score Aberrant Percentage: {zaberrantpercentage:.2f}%")

    print("="*40)
```

```python

zscore_threshold = 3
columnnames = [ 'ad_sk_id_curr_ct', 'ad_amt_annuity_ct',
       'ad_amt_application_ct', 'ad_amt_credit_ct', 'ad_amt_down_payment_ct',
       'ad_amt_goods_price_ct', 'ad_hour_appr_process_start_ct',
       'ad_nflag_last_appl_in_day_ct', 'ad_rate_down_payment_ct',
       'ad_days_decision_ct', 'ad_sellerplace_area_ct', 'ad_cnt_payment_ct']
for columnname in columnnames:
    values = previous_data1[columnname]

    if not pd.api.types.is_numeric_dtype(values):
        continue

    zscores = np.abs((values - values.mean()) / values.std())

    aberrantz = values[zscores > zscorethreshold]
    zaberrantpercentage = len(aberrantz) / len(values) * 100

    print(f"Column: {columnname}")
    print(f"Z-score Aberrant Percentage: {zaberrantpercentage:.2f}%")

    plt.figure(figsize=(6, 4))
    sns.boxplot(x=values)
    plt.title(f"Boxplot for {columnname}")
    plt.xlabel(columnname)
    plt.ylabel('Values')
    plt.show()

    print("="*40)
```
